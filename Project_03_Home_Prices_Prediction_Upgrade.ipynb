```python
#Split numeric columns and string columns
numeric_cols = house_df.select_dtypes(include = ["int64","float64"]).columns
categorical_cols = house_df.select_dtypes(include = ["object"]).columns
test_categorical_cols = test_df.select_dtypes(include = ["object"]).columns
numeric_cols = numeric_cols[:-1]
```

```python
#Processing NaN in numeric columns
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
house_df[numeric_cols] = imputer.fit_transform(house_df[numeric_cols])
test_df[numeric_cols] = imputer.transform(test_df[numeric_cols])
```

```python
#Processing NaN in categorical columns
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

house_ex_price_df = house_df.drop(columns = "SalePrice")

ct = ColumnTransformer(transformers = [("encoder", OneHotEncoder(sparse_output=False, handle_unknown = 'ignore'), categorical_cols)], remainder = "passthrough")
#house_df[categorical_cols] = ct.fit_transform(house_df[categorical_cols])
encoded_array = ct.fit_transform(house_ex_price_df)
encoded_cols = ct.get_feature_names_out()

test_array = ct.transform(test_df)

#encoded_cols
categorical_df = pd.DataFrame(encoded_array, columns = encoded_cols, index = house_df.index)
test_categorical_df = pd.DataFrame(test_array, columns = encoded_cols, index = test_df.index)
#categorical_df
numerical_df = house_ex_price_df.drop(columns = categorical_cols)
test_numerical_df = test_df.drop(columns = categorical_cols)

house_new_df = pd.concat([numerical_df, categorical_df], axis = 1)
test_new_df = pd.concat([test_numerical_df, test_categorical_df], axis = 1)

#ct
#house_ex_price_df.head()
house_new_df["SalePrice"] = house_df["SalePrice"]
house_new_df["SalePrice"] = house_new_df.pop("SalePrice")
house_new_df.head()
```
