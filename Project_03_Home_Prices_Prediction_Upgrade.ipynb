```python
#Split numeric columns and string columns
numeric_cols = house_df.select_dtypes(include = ["int64","float64"]).columns
categorical_cols = house_df.select_dtypes(include = ["object"]).columns
numeric_cols = numeric_cols[:-1]
```

```python
#Processing NaN in numeric columns
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
house_df[numeric_cols] = imputer.fit_transform(house_df[numeric_cols])
test_df[numeric_cols] = imputer.transform(test_df[numeric_cols])
```
