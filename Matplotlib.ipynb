```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot()
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), linestyle = "dashed", label = "sin(x)");
plt.plot(x, np.cos(x), label = "cos(x)")
plt.xlabel = ("x")
plt.ylabel = ("y")
plt.axis([0,4,-0.7,0.6])
plt.axis("equal")
plt.legend();
```

```python
x = [1,2,3,4]
y = [11,22,33,44]

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,y)
ax.set(title = "A simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis");
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.randn(100)
sizes = 100*rng.randn(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s=sizes, c=colors, cmap = "viridis", alpha = 1);
fig.colorbar(img1)
```

```python
soft_drink_prices = {"Coke": 10,
                    "Pepsi": 12,
                    "Sprite": 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title = "Winmart",
      ylabel = "Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(soft_drink_prices.keys(), soft_drink_prices.values());
```
