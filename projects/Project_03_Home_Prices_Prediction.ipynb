```python
import numpy as np
import pandas as pd
```

```python
house_df = pd.read_csv("C:/Users/HELLO/Desktop/Data Science/data/house_prices_data_for_ml/train.csv", index_col = "Id")
```

```python
house_df.head()
```

```python
features = ["MSSubClass", "LotFrontage", "LotArea", "PoolArea", "MiscVal", "MoSold", "YrSold", "YearBuilt", "1stFlrSF", "2ndFlrSF", "FullBath", "TotRmsAbvGrd"]
```

```python
x = house_df[features]
```

```python
y = house_df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x,y,train_size = 0.8, test_size = 0.2, random_state = 0)
```

```python
x_train.shape
```

```python
x_valid.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state = 1)
```

```python
dt_model.fit(x_train,y_train)
```

```python
dt_model.predict(x_valid)
```

```python
y_predis = dt_model.predict(x_valid.head())
```

```python
y_predis
```

```python
pd.DataFrame({"y": y_valid.head(), "y_predis": y_predis})
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
rf_model = RandomForestRegressor(random_state = 1)
rf_model.fit(x_train,y_train)
```

```python
rf_val_predis = rf_model.predict(x_valid)
```
