```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("C:/Users/HELLO/Desktop/Data Science/data/california_cities.csv")
cities.head(5)
```

```python
lat = cities['latd']
lon = cities['longd']

cities['area_total_km2'].sort_values(ascending = False)
```

```python
lat[:5]
```

```python
plt.figure(figsize = (8,6))
plt.style.use('seaborn-v0_8')
population, area = cities['population_total'], cities['area_total_km2']
plt.scatter(lon,lat, c=np.log10(population), s = area, cmap = 'viridis', linewidth = 0, alpha = 1);
plt.axis('equal');
plt.xlabel('longlatitude');
plt.ylabel('latitude');
plt.colorbar(label = 'log$_{10}$(population)');
plt.clim(3,7);

area_range = [50,100,300,500]

for area in area_range:
    plt.scatter([],[], s = area, label = str(area)+'km^2')

plt.legend(scatterpoints = 1, labelspacing = 1, title = 'City Area')

plt.title("California Cities: Population and Area Distribution");
```
