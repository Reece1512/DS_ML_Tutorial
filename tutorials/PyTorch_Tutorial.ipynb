### Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning:

1. Element-wise multiplication
2. Matrix multiplication (dot product)

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match:
* `(3,2) @ (3,2)` won't work
* `(2,3) @ (3,2)` will work
* `(3,2) @ (2,3)` will work

2. The resulting matrix has the shape of the **outer dimensions**:
* `(2,3) @ (3,2)` -> `(2,2)`
* `(3,2) @ (2,3)` -> `(3,3)`

```python
# Element wise multiplication
print(tensor, '*', tensor)
print(f"Equal {tensor*tensor}")
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

```python
%%time
value = 0
for i in range (0, len(tensor)):
    value += tensor[i]*tensor[i]
value
```

```python
%%time
torch.matmul(tensor,tensor)
```

### One of the most common errors in deep learning

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])
tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])
#torch.mm is the same as torch.matmul (it's an alies for writing less code)
torch.matmul(tensor_A, tensor_B) 
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our Tensor shape issues, we can manipulate the shape of one of our Tensors using a **transpose**

A **transpose** switches the axes or dimensions of a given Tensor.

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")

torch.matmul(tensor_A, tensor_B.T).shape
```

## Reshaping, stacking, squeezing, and unsqueezing Tensors

* Reshaping - reshapes an input Tensor to a defined shape
* View - Return a view of an input Tensor of certain shape but keep the same memory as the original Tenosr
* Stacking - combine multiple Tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - remove all '1' dimensions from a Tensor
* Unsqueeze - add a '1' dimension to a target Tensor
* Permute - return a view of the input with dimensions 

```python
# Let's create a Tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
# Change a view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because a view of a Tensor shares the same memory as the original input)
z[:, 0] = 5
z
```

```python
# Stack Tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim = 0)
x_stacked
```

```python
# torch.squeeze() - removes all single dimensions from a target Tensor
x_reshaped
```

```python
x_reshaped.shape
```

```python
x_squeezed = x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
# torch.unsqueeze() = adds a single dimension to a target tensor at a specific dim
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

x_unsqueezed = x_squeezed.unsqueeze(dim = 1)

print(f"New target: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimensions of a target Tensor in a specified order
x_original = torch.rand(size = (224, 224, 3)) #[height, width, color_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) #shift axis 0 -> 1, 1 -> 2, 2 -> 0

print(f"original: {x_original.shape}")
print(f"permuted: {x_permuted.shape}")
```
